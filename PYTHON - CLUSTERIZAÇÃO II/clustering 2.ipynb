```python
from sklearn.preprocessing import Normalizer
values = Normalizer().fit_transform(dados.values)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, n_init=10, max_iter=300)
y_pred = kmeans.fit_predict(values)
```

## Silhouette

#### Valor entre 0 e 1

```python
from sklearn import metrics
labels = kmeans.labels_
silhouette = metrics.silhouette_score(values,labels, metric='euclidean')
print(silhouette)
```

## Davies Boudin

#### Quanto mais próximo de 0 melhor

```python
dbs = metrics.davies_bouldin_score(values, labels)
print(dbs)
```

## Calinski Harabasz

####             Quanto mais alto o valor melhor

```python
calinski = metrics.calinski_harabasz_score(values, labels)
print(calinski)
```

## Validação

```python
# função para aplicar os algoritimos com diferentes número de cluesteres

def cluestering_algorithm(n_clusters, dataset):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, max_iter=300)
    labels = kmeans.fit_predict(dataset)
    s = metrics.silhouette_score(dataset, labels, metric='euclidean')
    dbs = metrics.davies_bouldin_score(dataset, labels)
    calinski = metrics.calinski_harabasz_score(dataset, labels)
    return s, dbs, calinski
    
```

```python
s1, dbs1, calisnki1 = cluestering_algorithm(3, values)
print(s1, dbs1, calisnki1)
```

```python
s2, dbs2, calisnki2 = cluestering_algorithm(5, values)
print(s2, dbs2, calisnki2)
```

```python
s3, dbs3, calisnki3 = cluestering_algorithm(20, values)
print(s3, dbs3, calisnki3 )
```

```python
s4, dbs4, calisnki4 = cluestering_algorithm(50, values)
print(s4, dbs4, calisnki4)
```

```python
# gerando dados aleatórios para validação
random_data = np.random.rand(8950, 16)
s5, dbs5, calisnki5 = cluestering_algorithm(5, random_data)
print(s5, dbs5, calisnki5)
print(s2, dbs2, calisnki2)
```

```python
# verificando estabilidade dos clusteres
set1, set2, set3 = np.array_split(values, 3)
s1, dbs1, calisnki1 = cluestering_algorithm(5, set1)
print(s1, dbs1, calisnki1)

s2, dbs2, calisnki2 = cluestering_algorithm(5, set2)
print(s2, dbs2, calisnki2)

s3, dbs3, calisnki3 = cluestering_algorithm(5, set3)
print(s3, dbs3, calisnki3)
```
