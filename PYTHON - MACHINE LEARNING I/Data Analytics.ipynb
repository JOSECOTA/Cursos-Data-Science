```python
import pandas as pd
```

```python
dados_treino = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/digital-marketing-ml/master/train.csv')
dados_treino.head()
```

```python
dados_treino.info()
```

```python
dados_treino.date = pd.to_datetime(dados_treino.date, format='%Y%m%d')
```

```python
dados_treino.head()
```

```python
dados_treino.fullVisitorId = dados_treino.fullVisitorId.astype(str)
dados_treino.visitId = dados_treino.visitId.astype(str)
```

```python
dados_treino.columns
```

```python
import json
```

```python
dicionarios = ['device', 'geoNetwork', 'trafficSource','totals']

for colunas in dicionarios:
    dados_treino = dados_treino.join(pd.DataFrame([json.loads(linha) for linha in dados_treino[colunas]]))
    
```

```python
dados_treino.head()
```

```python
dados_treino.drop(columns=dicionarios, axis=1, inplace=True)
```

```python
dados_treino.head()
```

```python
dados_treino.drop(columns= 'adwordsClickInfo', axis=1, inplace=True)
```

```python
coluna_na = []
for coluna in dados_treino.columns:
    print(coluna + ': ' + str(len(dados_treino[coluna].unique())))
    if len(dados_treino[coluna].unique()) == 1:
        coluna_na.append(coluna)
```

```python
dados_treino.drop(columns=coluna_na, axis= 1, inplace=True)
dados_treino.head()
```

```python
quant = ['bounces', 'hits', 'newVisits','pageviews','transactionRevenue']
```

```python
for coluna in quant:
    dados_treino[coluna] = pd.to_numeric(dados_treino[coluna])
```

```python
dados_treino.transactionRevenue = dados_treino.transactionRevenue.fillna(0)
```

```python
dados_treino.transactionRevenue = dados_treino.transactionRevenue / 1000000
```

```python
dados_treino.shape
```

```python
len(set(dados_treino.fullVisitorId))
```

```python
df_quant = dados_treino.groupby('fullVisitorId', as_index=False)[quant].sum()
```

```python
df_quant.head()
```

```python
y = df_quant.transactionRevenue.copy()
X = df_quant.drop('transactionRevenue', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
reg_predict = reg.predict(X_test)
reg_predict
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
```

```python
resultados[resultados.revenue > 0]
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
np.sqrt(mean_squared_error(y_test, reg_predict))
```
