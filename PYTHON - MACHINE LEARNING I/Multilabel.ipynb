```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/alura_classificacao_multilabel/master/dataset/stackoverflow_perguntas.csv')
```

```python
dados.head()
```

```python
dados.info()
```

```python
list_tags = dados.Tags.values.tolist()
```

```python
labels = list()
for tags in list_tags:
    for tag in tags.split():
        if tag not in labels:
            labels.append(tag)
    
```

```python
labels
```

```python
node_js = []
jquery = []
html = []
angular = []

for linha_tag in dados.Tags:
    if 'node.js' in linha_tag:
        node_js.append(1)
    else:
        node_js.append(0)

for linha_tag in dados.Tags:
    if 'jquery' in linha_tag:
        jquery.append(1)
    else:
        jquery.append(0)     

for linha_tag in dados.Tags:
    if 'html' in linha_tag:
        html.append(1)
    else:
        html.append(0)

for linha_tag in dados.Tags:
    if 'angular' in linha_tag:
        angular.append(1)
    else:
        angular.append(0)       
```

```python
dados['node.js'] = node_js
dados['jquery'] = jquery
dados['html'] = html
dados['angular'] = angular
```

```python
dados.drop(columns='Tags', axis=1, inplace=True)
dados.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
lista_zip = list(zip(dados['node.js'],dados['jquery'], dados['html'],dados['angular']))
```

```python
lista_zip
```

```python
dados['todas_tags'] = lista_zip
dados.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados.Perguntas, dados.todas_tags, test_size=0.2, random_state=123)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vetorizador = TfidfVectorizer(max_features=5000, max_df=0.85)
```

```python
vetorizador.fit(dados.Perguntas)
X_train_tfidf = vetorizador.transform(X_train)
X_test_tfidf = vetorizador.transform(X_test)
```

```python
X_train_tfidf.shape
```

```python
X_test_tfidf.shape
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
y_train_array = np.asarray(list(y_train))
y_test_array = np.asarray(list(y_test))
y_train_array
```

```python
regressao_logistica = LogisticRegression()
classificador_ovr = OneVsRestClassifier(regressao_logistica)
classificador_ovr.fit(X_train_tfidf, y_train_array)
resultado_ovr = classificador_ovr.score(X_test_tfidf, y_test_array) 
resultado_ovr
```

```python
from sklearn.metrics import hamming_loss
```

```python
y_pred = classificador_ovr.predict(X_test_tfidf)
hamming_loss(y_test_array,y_pred)
```
